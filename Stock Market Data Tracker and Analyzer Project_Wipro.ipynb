**Short Term:** 20 days

```python
#Import the finance package
import yfinance as yf
#import the matplotlib library for plotting
import matplotlib.pyplot as plt

#Define the ticker symbol for the stock to analyze
ticker_symbol = "WIT"
#Access stock information for the ticker
print("Wipro Limited Stock Data for 20 days - short term")
stock_data = yf.Ticker(ticker_symbol)
# Retrieve historical data for the stock
# Retrieve historical data for 20 days
start_date = "2024-02-01"
end_date = "2024-02-20"

hist_data = stock_data.history(start=start_date, end=end_date)
#Display the first few rows to understand the data structure
display(hist_data)
#calculate the daily returns, which is the percentage charge in closing price day-over-day
hist_data["Daily Return"] = hist_data['Close'].pct_change()
```

```python
#Ensure plots displayed inline in the notebook
%matplotlib inline

#Plot the closing price of the stock
plt.figure(figsize=(20,5))
plt.plot((hist_data['Close']), label = 'Wipro Limited close price, Short-term 20 days')
plt.title("Closing price over time")
plt.xlabel('Data')
plt.ylabel('Closing Price USD ($)')
plt.legend()

#calculate the maximum closing price and print it
max_price_date = hist_data['Close'].idxmax()
max_price = hist_data['Close'].max()
print(f"Maximum closing price: {max_price:.2f}")

#calculate the minimum closing price and print it
min_price_date = hist_data['Close'].idxmin()
min_price = hist_data['Close'].min()
print(f"Minimum closing price: {min_price:.2f}")

plt.scatter(min_price_date, min_price, color='red', label=f'Min Price: ${min_price:.2f}')
plt.scatter(max_price_date, max_price, color='red', label=f'Max Price: ${max_price:.2f}')

plt.annotate(f'Min Price: ${min_price:.2f}', (min_price_date, min_price), xytext=(-50, 30), textcoords='offset points',
             arrowprops=dict(arrowstyle="->", color='blue'))
plt.annotate(f'Max Price: ${max_price:.2f}', (max_price_date, max_price), xytext=(-50, -30), textcoords='offset points',
             arrowprops=dict(arrowstyle="->", color='green'))

plt.grid(True)
plt.show()
```

```python
#Calculate the percentage change from the first to the last clsoing price in the period
percentage_change = ((hist_data['Close'][-1] - hist_data['Close'][0]) / hist_data['Close'][0]) * 100
print(f"Wipro 20 days percentage change in closing price {percentage_change:.2f}%")
```
